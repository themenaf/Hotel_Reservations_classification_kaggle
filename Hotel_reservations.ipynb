```python
dataFrame['type_of_meal_plan']=dataFrame['type_of_meal_plan'].astype('int')
dataFrame['room_type_reserved']=dataFrame['room_type_reserved'].astype('int')
dataFrame['market_segment_type']=dataFrame['market_segment_type'].astype('int')
dataFrame['booking_status']=dataFrame['booking_status'].astype('int')
```

```python
dataFrame= dataFrame.astype('int')
```

```python
y= dataFrame["booking_status"].values
X=dataFrame.drop("booking_status",axis=1).values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X= scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=100)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
